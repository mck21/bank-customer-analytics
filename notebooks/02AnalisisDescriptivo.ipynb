## 3. Análisis descriptivo de los datos

Carga de datos limpios

```python
from IPython.display import display

import sys
sys.path.append('../src')
from data.data_loader import load_clean_data

# Cargar datos limpios
df_bank = load_clean_data()
display(df_bank.head())
```

### 3.1 Análisis de variables numéricas

```python
# Análisis descriptivo básico
print("\nInformación general del dataset:")
print(df_bank.info())

print("\nEstadísticas descriptivas de variables numéricas:")
print(df_bank.describe())
```

### 3.2 Análisis de variables categóricas

```python
print("\nDistribución de variables categóricas:")
categorical_cols = df_bank.select_dtypes(include=['object']).columns
for col in categorical_cols:
    print(f"\nDistribución de {col}:")
    print(df_bank[col].value_counts())
    porcentajes = df_bank[col].value_counts(normalize=True) * 100
    print(f"Porcentaje:\n{porcentajes.round(2).astype(str) + '%'}")
```

### 3.3 Análisis de correlaciones

```python
# Matriz de correlación
correlation_matrix = df_bank.select_dtypes(include=['int64','float64']).corr()
print("\nMatriz de correlación:")
print(correlation_matrix)

# Identificar correlaciones más fuertes (>0.5 o <-0.5)
strong_correlations = []
for i in range(len(correlation_matrix.columns)):
    for j in range(i):
        if abs(correlation_matrix.iloc[i,j]) > 0.5:
            strong_correlations.append((
                correlation_matrix.index[i], 
                correlation_matrix.columns[j], 
                correlation_matrix.iloc[i,j]
            ))

print("\nCorrelaciones más fuertes (>0.5 o <-0.5):")
for var1, var2, corr in strong_correlations:
    print(f"{var1} - {var2}: {corr:.3f}")
```

### 3.4 Análisis de outliers


```python
# Análisis de outliers mediante estadísticas
numeric_cols = df_bank.select_dtypes(include=['int64','float64']).columns
for col in numeric_cols:
    Q1 = df_bank[col].quantile(0.25)
    Q3 = df_bank[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df_bank[(df_bank[col] < lower_bound) | (df_bank[col] > upper_bound)][col]
    
    print(f"\nAnálisis de outliers para {col}:")
    print(f"Número de outliers: {len(outliers)}")
    print(f"Porcentaje de outliers: {(len(outliers)/len(df_bank)*100):.2f}%")
    print(f"Rango de outliers: [{outliers.min():.2f}, {outliers.max():.2f}]")
```
